```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('all_data.csv')
print(df.head())
```

```python
country_val = df.Country.value_counts()
print(country_val)
```

```python
years_val = df['Life expectancy at birth (years)']
print(years_val)
```

```python
life_exp_eeuu = df['Life expectancy at birth (years)'][df.Country == 'United States of America']
print(life_exp_eeuu)

print(np.mean(life_exp_eeuu))

# EEUU: 78.06 years.
```

```python
life_exp_chile = df['Life expectancy at birth (years)'][df.Country == 'Chile']
print(life_exp_chile)

print(np.mean(life_exp_chile))

# Chile: 78.94 years
```

```python
life_exp_germany = df['Life expectancy at birth (years)'][df.Country == 'Germany']
print(life_exp_germany)

print(np.mean(life_exp_germany))

# Germany: 79.65 years
```

```python
life_exp_mexico = df['Life expectancy at birth (years)'][df.Country == 'Mexico']
print(life_exp_mexico)

print(np.mean(life_exp_mexico))
# Mexico: 75.72 years
```

```python
life_exp_china = df['Life expectancy at birth (years)'][df.Country == 'China']
print(life_exp_china)

print(np.mean(life_exp_china))

# China: 74.27 years
```

```python
life_exp_zimbabwe = df['Life expectancy at birth (years)'][df.Country == 'Zimbabwe']
print(life_exp_zimbabwe)

print(np.mean(life_exp_zimbabwe))

# Zimbabwe: 50.09 years
```

```python
years_min = np.min(years_val)
years_av = np.mean(years_val)
years_max = np.max(years_val)

print(years_min)
print(years_av)
print(years_max)
```

```python
sns.scatterplot(x='Life expectancy at birth (years)', y='GDP', data=df, hue = 'Country')
plt.show()
plt.clf()

# The higher the GDP of a country the higher the life expectancy will be.
```

```python
sns.scatterplot(x='Year', y='Life expectancy at birth (years)', data=df, hue = 'Country')
plt.show()
plt.clf()

# We can see an upward trend at Life Expectancy over the years.
```

```python
sns.scatterplot(x='Year', y='GDP', data=df, hue = 'Country', marker = 's')
plt.legend(loc = 'upper left')
plt.show()
plt.clf()

# We can see that over the years, GDP of each country has increased considerably
```
